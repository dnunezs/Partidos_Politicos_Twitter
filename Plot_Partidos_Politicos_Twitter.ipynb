```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

Vox_Twitter = pd.read_csv('Vox_Twitter.csv', sep=',', index_col=0)
Podemos_Twitter = pd.read_csv('Podemos_Twitter.csv', sep=',', index_col=0)
PP_Twitter = pd.read_csv('PP_Twitter.csv', sep=',', index_col=0)
PSOE_Twitter = pd.read_csv('PSOE_Twitter.csv', sep=',', index_col=0)
```

```python
#Hacemos un grafico de los datos de seguidores y seguidos
#Preparamos las etiquetas de x
meses = ['Junio 2017', 'Enero 2018', 'Junio 2018','Octubre', 'Enero 2019', 'Junio 2019', 'Enero 2020', 'Junio 2020'  ]
mapeado = (2,32,54,64,85, 106, 137, 158)

#Hacemos el grafico
fig, axs2 = plt.subplots(nrows=2, ncols=1, figsize=(20,10))
axs2[0].plot(np.arange(len(Vox_Twitter.index)), pd.to_numeric(Vox_Twitter.Seguidores), color = 'g')
axs2[0].set_title("Evolucion de Seguidores de Vox en Twitter")
axs2[0].axvline(x=64,linewidth=2, color='black')
axs2[1].plot(np.arange(len(Vox_Twitter.index)), pd.to_numeric(Vox_Twitter.Seguidos), color = 'b')
axs2[1].set_title("Evolucion de Seguidos por Vox en Twitter")
plt.xticks(mapeado, meses)
plt.savefig('pct_tw_vox.png', bbox_inches='tight')
plt.show()
```

```python
# Analisis y respuestas
meses = ['Bannon A18', 'Andalucia D18', 'Gen', 'Aut', 'Generales N19'  ]
mapeado = (46,79, 100,104, 128 )

plt.figure(figsize=(20,4))

plt.plot(np.arange(len(Vox_Twitter.index)), pd.to_numeric(Vox_Twitter.Seguidores), color = 'g')
plt.axvline(x=46,linewidth=2, color='b', alpha =0.5)
plt.axvline(x=79,linewidth=2, color='b',  alpha =0.5)
plt.axvline(x=100,linewidth=2, color='b',  alpha =0.5)
plt.axvline(x=128,linewidth=2, color='b',  alpha =0.5)
plt.axvline(x=104,linewidth=2, color='b',  alpha =0.5)
plt.xticks(mapeado, meses)
plt.show()
```

```python
# Analisis y respuestas de los nuevos seguidores
meses = ['2017','2018', 'Bannon', 'AND', '2019', 'Gen', 'Aut', 'Generales','2020']
mapeado = (0,32, 46,79, 85, 100,104, 128, 137 )

PSOE_Twitter = PSOE_Twitter[48:]

plt.figure(figsize=(20,4))

plt.plot(np.arange(len(Vox_Twitter.index)), pd.to_numeric(Vox_Twitter.Seguidores), color = 'g', label="Vox")
plt.plot(np.arange(len(Podemos_Twitter.index)), pd.to_numeric(Podemos_Twitter.Seguidores), color = 'purple', label="Podemos")
plt.plot(np.arange(len(PP_Twitter.index)), pd.to_numeric(PP_Twitter.Seguidores), color = 'b', label="PP")
plt.plot(np.arange(len(PSOE_Twitter.index)), pd.to_numeric(PSOE_Twitter.Seguidores), color = 'r', label="PSOE")

plt.axvline(x=46,linewidth=2, color='b', alpha =0.2)
plt.axvline(x=79,linewidth=2, color='b',  alpha =0.2)
plt.axvline(x=100,linewidth=2, color='b',  alpha =0.2)
plt.axvline(x=128,linewidth=2, color='b',  alpha =0.2)
plt.axvline(x=104,linewidth=2, color='b',  alpha =0.2)
plt.axhline(0,linewidth=2, color='black')
plt.xticks(mapeado, meses)
plt.ylim(-2000, 15000)
plt.legend()
plt.show()
```
